```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('cars.csv', sep=';')
```

```python
dataset['price_usd'] = [x.replace(',', '.') for x in dataset['price_usd']]
dataset['price_usd'] = dataset['price_usd'].astype('float64')
```

```python
dataset['vol'] = [x.replace(',', '.') for x in dataset['vol']]
dataset['vol'] = dataset['vol'].astype('float64')
```
